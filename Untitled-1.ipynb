```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.cluster import KMeans
```

```python
img = cv2.imread('img1.jpg')
print(img.shape)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
X = img.reshape(-1, 3)
X.shape
```

```python
cluster = 10
kmeans = KMeans(n_clusters=cluster, init="random", n_init=10).fit(X)
print(kmeans)

centers = kmeans.cluster_centers_
labels = kmeans.labels_
print(centers)
print(centers.shape)

seg_img = centers[labels]
seg_img = np.reshape(seg_img, img.shape)
plt.imshow(seg_img.astype('uint8'))
plt.show()
```

```python
fix, axes = plt.subplots(2, 5, figsize=(12, 8))
for ax, i in zip(axes.flatten(), range(10)):
    clus_cen = centers.copy()
    clus_cen[np.arange(clus_cen.shape[0]) != i] = 255
    clus_img = clus_cen[labels]
    clus_img = np.reshape(clus_img, img.shape)
    ax.imshow(clus_img.astype('uint8'))
    ax.axis('off')

plt.tight_layout()
plt.show()
```
